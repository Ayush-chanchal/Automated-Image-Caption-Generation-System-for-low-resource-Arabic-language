```python
import pandas as pd
from googletrans import Translator
from requests.exceptions import Timeout
```

```python
file_path = './MAIN_captions.xlsx'
df = pd.read_excel(file_path)
translator = Translator()
def translate_to_arabic(text):
    try:
        translation = translator.translate(text, dest='ar')
        return translation.text
    except Timeout as e:
        print(f"Translation timed out for: {text}")
        
# Apply translation to each caption in the dataset
df['Arabic_Caption'] = df['caption'].apply(translate_to_arabic)

# Save the translated dataset
output_path = 'translated_dataset.xlsx'
df.to_excel(output_path, index=False)

print(f'Translation completed. Translated dataset saved to {output_path}')
```
